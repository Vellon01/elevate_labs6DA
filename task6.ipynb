```python
import pandas as pd
import sqlite3
```

```python
df=pd.read_csv("sales_data_sample.csv")
```

```python
conn = sqlite3.connect('sales_data.db')
df.to_sql("sales_data_table",conn,if_exists="replace",index=False)
```

```python
pd.read_sql_query('PRAGMA table_info(sales_data_table);', conn)
```

```python
pd.read_sql_query("SELECT strftime('%m', ORDERDATE) AS order_month, count(*) AS total_orders FROM sales_data_table GROUP BY MONTH_ID", conn)
```

```python
pd.read_sql_query("SELECT sum(SALES) AS total_revenue FROM sales_data_table ", conn)
```

```python
pd.read_sql_query("SELECT count(ORDERNUMBER) AS total_orders FROM sales_data_table", conn)
```

```python
pd.read_sql_query("SELECT CITY,STATE,COUNTRY FROM sales_data_table ORDER BY COUNTRY", conn)
```

```python
pd.read_sql_query("SELECT * FROM sales_data_table WHERE ORDERDATE BETWEEN '2003-01-01' AND '2003-07-31' GROUP BY YEAR_ID,MONTH_ID", conn)
```

```python
pd.read_sql_query("SELECT YEAR_ID || '-' || printf('%02d', MONTH_ID) AS Month, COUNT(DISTINCT ORDERNUMBER) AS Total_Orders, ROUND(SUM(SALES), 2) AS Total_Revenue FROM sales_data_table GROUP BY YEAR_ID, MONTH_ID ORDER BY YEAR_ID, MONTH_ID", conn)
```
